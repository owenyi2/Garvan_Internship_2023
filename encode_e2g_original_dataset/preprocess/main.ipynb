```python
import pandas as pd
```

```python
df = pd.read_csv("EPCrisprBenchmark_ensemble_data_GRCh38.K562_ENCDO000AAD_ENCFF325RTP_DNaseOnly_features_NAfilled.tsv", sep="\t")
df
```

# Issue 1: C19orf43 of the FlowFISH_K562 dataset has the TSS missing

```python
df[df[["startTSS", "endTSS"]].isna().any(axis=1)]
```

```python
# Thankfully the same gene C19orf43 occurs in the Gasperini2019 dataset with labeled TSS position
df[df["name"].str.contains("C19orf43")]
```

```python
# impute missing C19orf43 with Gasperini dataset

df.loc[df[["startTSS", "endTSS"]].isna().any(axis=1), "startTSS"] = 12734774
df.loc[df[["startTSS", "endTSS"]].isna().any(axis=1), "endTSS"] = 12734775 #impute based on Gasperini dataset
```

# Issue 2: One row has missing value in `Regulated` column
`Regulated` is a boolean describing if candidate pair is a genuine EPI

```python
df = df.drop(df[df["Regulated"].isna()].index) 
```

```python
df.isna().any(axis=0)
```

```python
df.to_csv("EPCrisprBenchmark_ensemble_data_GRCh38_imputed_TSS_filter_NaN_regulated.tsv", sep="\t", index=None)
```
